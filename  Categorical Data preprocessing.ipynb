## IMPLEMENTATION On MULTIPLE DATA PRE-PROCESSING METHODS AT A TIME :

- use sklearn column transformer 
- we have to transform X_train and X_test

```python
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import OrdinalEncoder,MinMaxScaler
```

```python
tf=ColumnTransformer(transformers=[("t1",OrdinalEncoder(),[0,1]),
                                   ("t2",MinMaxScaler(),[2,3,4,5,6])])
X_train_trans=tf.fit_transform(X_train)
X_test_trans=tf.transform(X_test)
```

```python
X_train.columns
```

```python
X_train_trdf=pd.DataFrame(X_train_trans,columns=X_train.columns)
X_train_trdf.head()
```
